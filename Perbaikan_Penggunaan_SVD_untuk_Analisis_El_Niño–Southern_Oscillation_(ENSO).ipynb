<h1><b>Teori Singular Value Decomposition (SVD)</b></h1>

Misalkan $A \in \mathbb{R}^{m \times n}$ adalah suatu matriks real.  
Singular Value Decomposition menyatakan bahwa matriks $A$ dapat diuraikan menjadi

$$
A = U \Sigma V^T
$$

dengan:
- $U \in \mathbb{R}^{m \times m}$ adalah matriks ortogonal,
- $V \in \mathbb{R}^{n \times n}$ adalah matriks ortogonal,
- $\Sigma \in \mathbb{R}^{m \times n}$ adalah matriks diagonal yang memuat nilai singular.

Elemen diagonal matriks $\Sigma$, yang disebut nilai singular, diurutkan sebagai

$$
\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0, \qquad r = \text{rank}(A)
$$


Dekomposisi ini memungkinkan matriks $A$ dituliskan sebagai penjumlahan komponen-komponen rank-satu:

$$
A = \sum_{i=1}^{r} \sigma_i u_i v_i^T
$$

di mana setiap pasangan $(u_i, v_i)$ merepresentasikan satu mode atau pola tertentu dalam data.

Pendekatan satu-mode diperoleh dengan hanya mempertahankan komponen dominan,

$$
A \approx \sigma_1 u_1 v_1^T
$$

yang sering kali sudah mampu menangkap struktur utama data.

Kontribusi energi dari mode ke-$i$ didefinisikan sebagai

$$
E_i = \frac{\sigma_i^2}{\sum_{j=1}^{r} \sigma_j^2}
$$

Mode dengan nilai $\sigma_i$ terbesar disebut sebagai mode dominan karena memberikan kontribusi terbesar terhadap variasi data.

<h1><b>Aplikasi SVD untuk Analisis El Ni√±o‚ÄìSouthern Oscillation (ENSO)</b></h1>
<p align="justify">
Fenomena El Ni√±o‚ÄìSouthern Oscillation (ENSO) pada dasarnya merupakan perubahan suhu permukaan laut yang terjadi secara luas di Samudra Pasifik dan berkembang dari waktu ke waktu. Karena perubahan ini tidak terjadi di satu titik saja, melainkan melibatkan banyak lokasi dan waktu, maka ENSO secara alami merupakan fenomena spasial‚Äìtemporal. Untuk dapat menganalisis fenomena ini secara matematis, langkah pertama yang dilakukan adalah merepresentasikan suhu permukaan laut sebagai fungsi ruang dan waktu:
$$
T(x,y,t)
$$
dengan $x$ dan $y$ menyatakan koordinat spasial (bujur dan lintang),
serta $t$ menyatakan waktu.
</p>
<hr>

<h2><b> Pembentukan Anomali Suhu Permukaan Laut</b></h2>
<p align="justify">
Data suhu permukaan laut kemudian diubah menjadi anomali dengan cara mengurangi nilai rata-rata klimatologisnya.
$$
T'(x,y,t) = T(x,y,t) - \overline{T}(x,y)
$$
dengan $\overline{T}(x,y)$ adalah nilai rata-rata klimatologis.
</p>
<p align="justify">
Langkah ini memiliki dasar matematis yang jelas, yaitu untuk menghilangkan komponen musiman yang bersifat periodik dan fokus pada fluktuasi yang lebih besar serta tidak rutin, yang diasosiasikan dengan ENSO. Dengan demikian, sinyal yang dianalisis benar-benar merepresentasikan variabilitas ENSO, bukan variasi tahunan biasa.
</p>
<hr>

<h2><b> Pembentukan Matriks Data</b></h2>
<p align="justify">
Setelah diperoleh data anomali, data tersebut didiskretisasi dan disusun ke dalam bentuk matriks. Setiap baris matriks merepresentasikan satu lokasi di Samudra Pasifik, sedangkan setiap kolom merepresentasikan satu waktu pengamatan.
$$
X \in \mathbb{R}^{m \times n}
$$
dengan:
</p>
<ul>
<li> $m$ menyatakan jumlah titik grid spasial,</li>
<li> $n$ menyatakan jumlah waktu pengamatan.</li>
</ul>
<p align="justify">
Setiap elemen matriks didefinisikan sebagai:
$$
X_{ij} = T'(x_i, y_i, t_j)
$$
Penyusunan ini penting karena memungkinkan fenomena fisik ENSO dipelajari dalam kerangka aljabar linear, yang menyediakan alat matematis yang kuat untuk analisis data berdimensi besar.
</p>
<hr>

<h2><b>Singular Value Decomposition (SVD)</b></h2>
<p align="justify">
Matriks data anomali kemudian diuraikan menggunakan Singular Value Decomposition (SVD). Secara matematis, SVD memisahkan matriks data menjadi komponen-komponen ortogonal yang diurutkan berdasarkan kontribusi variasinya. Dalam konteks ENSO, dekomposisi ini bertujuan untuk memisahkan pola dominan yang menjelaskan sebagian besar variabilitas suhu permukaan laut dari pola-pola yang kontribusinya lebih kecil.
$$
X = U \Sigma V^T
$$
dengan:
</p>
<ul align="justify">
<li> $U$ merepresentasikan pola spasial,</li>
<li> $\Sigma$ berisi nilai singular,</li>
<li> $V$ merepresentasikan pola temporal.</li>
</ul>
<p align="justify">
Nilai singular diurutkan sebagai:
$$
\sigma_1 \ge \sigma_2 \ge \cdots \ge \sigma_r > 0
$$
</p>
<hr>

<h2><b> Kontribusi Energi Mode ENSO</b></h2>
<p align="justify">
Kontribusi energi mode ke-$i$ didefinisikan sebagai:
$$
E_i = \frac{\sigma_i^2}{\sum_{j=1}^{r} \sigma_j^2}
$$
Mode pertama dengan nila $\sigma_1$ terbesar menjelaskan variasi terbesar dalam data dan secara matematis diinterpretasikan sebagai representasi dominan fenomena ENSO. Hasil ini menjadi dasar untuk melakukan pendekatan menggunakan satu mode utama.
</p>
<hr>

<h2><b> Pendekatan Mode Dominan </b></h2>
<p align="justify">
Berdasarkan hasil kontribusi energi, pendekatan menggunakan mode pertama dilakukan untuk merepresentasikan struktur dominan dalam data ENSO. Mode ini kemudian diinterpretasikan sebagai representasi utama ENSO, karena mampu menangkap pola spasial dan temporal yang konsisten dengan karakteristik El Ni√±o dan La Ni√±a.
</p>
<p align="justify">
Matriks $X$ dapat dituliskan sebagai penjumlahan komponen rank-satu:
$$
X = \sum_{i=1}^{r} \sigma_i u_i v_i^T
$$
Pendekatan satu-mode diberikan oleh:
$$
X \approx \sigma_1 u_1 v_1^T
$$
Pendekatan ini cukup untuk menangkap struktur utama ENSO
karena mode pertama memuat sebagian besar variabilitas data.
</p>

<h1><b>Simulasi Data Permukaan Laut</b></h1>
<p align="justify">
Untuk keperluan simulasi, digunakan data Extended Reconstructed Sea Surface Temperature versi 5 (ERSSTv5) dari NOAA, yaitu dataset global suhu permukaan laut. Analisis difokuskan pada wilayah Ni√±o 3.4 di Samudra Pasifik tengah, yang secara klimatologis merepresentasikan pusat interaksi laut‚Äìatmosfer paling stabil dan memiliki korelasi kuat dengan variabilitas ENSO, sehingga secara luas digunakan sebagai indikator standar El Ni√±o dan La Ni√±a oleh lembaga klimatologi internasional.
</p>
<p align="justify">
Rentang waktu pengamatan dibatasi pada periode 2014‚Äì2024, yang mencakup beberapa kejadian ENSO signifikan, termasuk El Ni√±o kuat pada 2015-2016, La Ni√±a berkepanjangan pada sekitar 2020-2022, serta El Ni√±o kembali pada 2023‚Äì2024. Untuk menghilangkan pengaruh siklus musiman, terlebih dahulu dihitung klimatologi bulanan dan kemudian dibentuk anomali suhu permukaan laut. Data anomali ini selanjutnya disusun ke dalam bentuk matriks spasio-temporal sebagai dasar penerapan metode Singular Value Decomposition (SVD) untuk mengidentifikasi pola dominan ENSO.
</p>

```python
import xarray as xr

# URL alternatif ke file NOAA ERSST v5
url = "https://downloads.psl.noaa.gov/Datasets/noaa.ersst.v5/sst.mnmean.nc"

# buka dataset langsung dari URL
ds = xr.open_dataset(url)

print(ds)

# Ambil variabel SST
sst = ds['sst']

# Potong data 2014‚Äì2024
sst_2014_2024 = sst.sel(time=slice("2014-01-01", "2024-12-31"))

print("Ukuran data 2014‚Äì2024:", sst_2014_2024.shape)
sst_2014_2024
```

```python
# ============================================
# SVD Mode-1 ENSO untuk periode 2014-2024
# ============================================

# -------------------------
# Imports
# -------------------------
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
from scipy.linalg import svd
from scipy.fft import fft, fftfreq
import warnings
warnings.filterwarnings("ignore")
plt.rcParams.update({'figure.max_open_warning': 0})

# -------------------------
# 1) Gunakan variabel sst_2014_2024 dari sel sebelumnya
# -------------------------
sst = sst_2014_2024.copy() # Pastikan kita bekerja dengan salinan jika ada modifikasi

print("Variabel SST shape (time, lat, lon) dari sel sebelumnya:", sst.shape)

# -------------------------
# 2) Pastikan lon 0..360
# -------------------------
if sst.lon.max() <= 180:
    sst = sst.assign_coords(lon=((sst.lon + 360) % 360)).sortby('lon')
    print("Longitude dikonversi ke 0..360")

# -------------------------
# 3) Pilih region NINO3.4: lat -5..5, lon 190..240
# -------------------------
region = sst.sel(lat=slice(5, -5), lon=slice(190, 240))
print("Region (time, lat, lon):", region.shape)

# -------------------------
# 4) Hitung anomali bulanan terhadap climatology periode ini
# -------------------------
climo = region.groupby('time.month').mean('time', skipna=True)
anom = region.groupby('time.month') - climo  # retains time coord
# Pastikan bentuk (lat, lon, time)
anom = anom.transpose('lat','lon','time')
lat = anom.lat.values
lon = anom.lon.values
times = anom.time.values
nlat = len(lat); nlon = len(lon); ntime = len(times)
print("Anomali shape (lat, lon, time):", anom.shape)

# -------------------------
# 5) Bentuk matriks space x time dan mask NaN
# -------------------------
X = anom.values.reshape(nlat*nlon, ntime)
mask_nan = np.any(np.isnan(X), axis=1)
X_masked = X[~mask_nan, :]
print("Matrix untuk SVD shape (space x time):", X_masked.shape)

# Cek apakah ada data
if X_masked.size == 0:
    raise RuntimeError("Tidak ada data valid setelah masking NaN. Periksa area/perioda.")

# -------------------------
# 6) SVD
# -------------------------
U, S, VT = svd(X_masked, full_matrices=False)
energy = (S**2)/np.sum(S**2)
print("SVD selesai. Mode count:", len(S))

# -------------------------
# Loop untuk menampilkan beberapa mode teratas
# -------------------------
num_modes_to_display = 3 # Tampilkan 3 mode teratas

for mode_idx in range(num_modes_to_display):
    current_mode_num = mode_idx + 1
    print(f"\n--- Menganalisis Mode {current_mode_num} ---")
    print(f"Kontribusi energi Mode-{current_mode_num} = {energy[mode_idx]*100:.2f}%")

    # Ambil komponen Mode-X (spasial & temporal)
    U_mode = U[:, mode_idx]    # spatial flattened (valid points)
    V_mode = VT[mode_idx, :]   # temporal coefficients (length ntime)
    S_value = S[mode_idx]

    U_grid = np.full(nlat*nlon, np.nan)
    U_grid[~mask_nan] = U_mode
    U_grid = U_grid.reshape(nlat, nlon)

    # Untuk interpretasi, seringkali V_mode * S_value memberi skala asli
    V_scaled = V_mode * S_value

    # -------------------------
    # 8) Plot: spatial map, temporal series, spectrum, energy bar
    # -------------------------
    fig = plt.figure(figsize=(14,10))

    ax1 = fig.add_subplot(2,2,1)
    c1 = ax1.contourf(lon, lat, U_grid, cmap='RdBu_r', levels=21)
    ax1.set_title(f"Mode-{current_mode_num} Spatial Pattern (NINO3.4 region)")
    ax1.set_xlabel("Longitude")
    ax1.set_ylabel("Latitude")
    plt.colorbar(c1, ax=ax1, orientation='vertical', label='Amplitude')

    ax2 = fig.add_subplot(2,2,2)
    ax2.plot(times, V_scaled, color='darkred', label=f'V{current_mode_num} * S{current_mode_num} (scaled)')
    ax2.plot(times, V_mode, color='k', alpha=0.6, label=f'V{current_mode_num} (unit norm)')
    ax2.set_title(f"Mode-{current_mode_num} Temporal Coefficient")
    ax2.set_xlabel("Time")
    ax2.set_ylabel("Amplitude")
    ax2.legend()
    ax2.grid(True)
    plt.xticks(rotation=25)

    ax3 = fig.add_subplot(2,2,3)
    # FFT on V_mode (demean)
    v_demean = V_mode - np.mean(V_mode)
    N_fft = len(v_demean)
    dt = 1/12.0  # years per sample (monthly)
    freqs = fftfreq(N_fft, d=dt)
    fft_power = np.abs(fft(v_demean))**2
    pos = freqs > 0
    periods = 1 / freqs[pos]
    ax3.plot(periods, fft_power[pos])
    ax3.set_xlim(0, 10)
    ax3.set_xlabel("Period (years)")
    ax3.set_ylabel("Power")
    ax3.set_title(f"Spectrum (Mode-{current_mode_num})")
    ax3.grid(True)

    ax4 = fig.add_subplot(2,2,4)
    modes_to_plot_bar = min(8, len(S))
    ax4.bar(np.arange(1, modes_to_plot_bar+1), energy[:modes_to_plot_bar]*100)
    ax4.set_xlabel("Mode")
    ax4.set_ylabel("Energy (%)")
    ax4.set_title("Energy contribution of top modes")
    ax4.grid(True)

    plt.tight_layout()
    plt.show()

    # -------------------------
    # 9) Simpan hasil (gambar & CSV) ke working dir (Colab)
    # -------------------------
    out_png = f"mode{current_mode_num}_summary_2014_2024.png"
    fig.savefig(out_png, dpi=160, bbox_inches='tight')
    print(f"Saved figure -> {out_png}")

    # Save time series Mode-X (scaled) to CSV
    df_ts = pd.DataFrame({'time': pd.to_datetime(times), f'V{current_mode_num}': V_mode, f'V{current_mode_num}_scaled': V_scaled})
    csv_ts = f"mode{current_mode_num}_timeseries_2014_2024.csv"
    df_ts.to_csv(csv_ts, index=False)
    print(f"Saved time series -> {csv_ts}")

    # Save spatial pattern to CSV (lat, lon, U_mode)
    csv_sp = f"mode{current_mode_num}_spatial_2014_2024.csv"
    rows = []
    for i_lat in range(nlat):
        for j_lon in range(nlon):
            rows.append((lat[i_lat], lon[j_lon], float(U_grid[i_lat, j_lon]) if not np.isnan(U_grid[i_lat,j_lon]) else np.nan))
    df_sp = pd.DataFrame(rows, columns=['lat','lon',f'U{current_mode_num}'])
    df_sp.to_csv(csv_sp, index=False)
    print(f"Saved spatial pattern -> {csv_sp}")

# -------------------------
# 10) Print ringkasan energi top modes (setelah loop)
# -------------------------
print("\nTop modes energy contribution (Summary):")
for i in range(min(10, len(S))):
    print(f"Mode {i+1}: {energy[i]*100:.2f}%")
```

</p>

```python
# ============================================================
# üåé Simulasi Aplikasi SVD untuk Analisis ENSO (Data Buatan)
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from scipy.fft import fft, fftfreq

# ------------------------------------------------------------
# 1Ô∏è‚É£ Membuat Data SST Buatan
# ------------------------------------------------------------
# Misal 20 tahun data bulanan ‚Üí 240 titik waktu
t = np.arange(0, 20, 1/12)
n_time = len(t)

# Buat grid spasial sederhana: 10x20 titik (misal koordinat Pasifik)
lon = np.linspace(120, 280, 20)  # bujur (longitude)
lat = np.linspace(-10, 10, 10)   # lintang (latitude)
Lon, Lat = np.meshgrid(lon, lat)

# Pola ENSO: pemanasan di Pasifik Tengah
enso_pattern = np.exp(-((Lon-220)**2)/500)  # pola spasial Gaussian di tengah

# Sinyal ENSO temporal: osilasi 5 tahunan
enso_time = np.sin(2*np.pi*t/5)

# Gabungkan pola spasial dan temporal jadi data 3D SST (lat √ó lon √ó waktu)
sst = np.zeros((len(lat), len(lon), n_time))
for i in range(n_time):
    sst[:, :, i] = 0.8 * enso_pattern * enso_time[i] + 0.2 * np.random.randn(*enso_pattern.shape)

# Visualisasi sebagian data SST (pada waktu tertentu)
plt.figure(figsize=(8,4))
plt.contourf(Lon, Lat, sst[:, :, 50], cmap='coolwarm')
plt.colorbar(label='Anomali SST (¬∞C)')
plt.title('Contoh Data SST Buatan (t = 20)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# ------------------------------------------------------------
# 2Ô∏è‚É£ Persiapkan Matriks untuk SVD
# ------------------------------------------------------------
# Ubah data 3D menjadi 2D (spasial √ó waktu)
X = sst.reshape(len(lat)*len(lon), n_time)

# Terapkan SVD
U, S, VT = svd(X, full_matrices=False)

# Hitung kontribusi energi
energy = (S**2) / np.sum(S**2)

# ------------------------------------------------------------
# 3Ô∏è‚É£ Visualisasi Pola Spasial (U) dan Temporal (V) untuk beberapa mode
# ------------------------------------------------------------
num_modes_to_display = 3 # Jumlah mode yang ingin ditampilkan

for mode_idx in range(num_modes_to_display):
    current_mode_num = mode_idx + 1
    print(f"\n--- Menganalisis Mode Buatan {current_mode_num} ---")
    print(f"Kontribusi energi Mode Buatan-{current_mode_num} = {energy[mode_idx]*100:.2f}%")

    # Ambil komponen mode saat ini
    U_mode = U[:, mode_idx].reshape(len(lat), len(lon))
    V_mode = VT[mode_idx, :]
    S_value = S[mode_idx]

    # Skalakan V_mode dengan nilai singular untuk representasi yang lebih intuitif
    V_scaled = V_mode * S_value

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.contourf(Lon, Lat, U_mode, cmap='coolwarm')
    plt.title(f'Pola Spasial (Mode Buatan {current_mode_num})')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.colorbar(label='Intensitas')

    plt.subplot(1, 3, 2)
    plt.plot(t, V_scaled, color='darkred', label=f'V{current_mode_num} * S{current_mode_num} (scaled)')
    plt.plot(t, V_mode, color='k', alpha=0.6, label=f'V{current_mode_num} (unit norm)')
    plt.title(f'Pola Temporal (Mode Buatan {current_mode_num})')
    plt.xlabel('Tahun')
    plt.ylabel('Amplitudo')
    plt.grid(True)
    plt.legend()

    # ------------------------------------------------------------
    # 4Ô∏è‚É£ Analisis Frekuensi (untuk melihat periode ENSO)
    # ------------------------------------------------------------
    plt.subplot(1, 3, 3)
    freqs = fftfreq(len(V_mode), 1/12)  # frekuensi dalam per tahun
    fft_power = np.abs(fft(V_mode - np.mean(V_mode)))**2
    # Filter frekuensi positif dan hitung periode
    pos = freqs > 0
    periods = 1 / freqs[pos]
    # Batasi plot periode ke 0-10 tahun
    plt.plot(periods, fft_power[pos])
    plt.xlabel('Periode (tahun)')
    plt.ylabel('Kekuatan Sinyal')
    plt.title(f'Spektrum Frekuensi (Mode Buatan {current_mode_num})')
    plt.grid(True)
    plt.xlim(0, 10)

    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 5Ô∏è‚É£ Tampilkan Kontribusi Energi Tiap Mode (Ringkasan)
# ------------------------------------------------------------
print("\nRingkasan Kontribusi Energi Mode Buatan:")
for i in range(min(10, len(S))): # Tampilkan hingga 10 mode teratas atau sebanyak yang tersedia
    print(f"Mode {i+1}: kontribusi energi = {energy[i]*100:.2f}%")
```

<h2> <b>Hasil dan Pembahasan</b></h2>

<h3><b>1. Nilai Singular ($\sigma_i$) dan Kontribusi Energi ($E_i$)</b></h3>
<p align="justify">
Dalam metode Singular Value Decomposition (SVD), matriks data $X$ diuraikan menjadi $U \Sigma V^T$, di mana nilai-nilai singular $\sigma_i$ menunjukkan seberapa besar peran masing-masing pola dalam menjelaskan variasi data. Hasil simulasi menunjukkan bahwa Mode 1 menyumbang 96.18% energi pada data NOAA dan 57.39% energi pada data buatan, yang berarti sebagian besar perubahan suhu permukaan laut di wilayah Ni√±o 3.4 dapat dijelaskan oleh satu pola utama ENSO. Secara sederhana, angka ini menunjukkan bahwa perubahan suhu laut tidak terjadi secara acak, melainkan didominasi oleh satu pola besar yang konsisten, sedangkan pola-pola lain hanya memberikan kontribusi kecil atau bersifat noise.
</p>
<p align="justify">
Perbedaan kontribusi energi Mode 1 antara data NOAA dan data buatan disebabkan oleh karakteristik data yang digunakan. Data NOAA difokuskan pada wilayah Ni√±o 3.4 dan telah diproses untuk menghilangkan variasi musiman, sehingga sinyal ENSO muncul sangat dominan. Sebaliknya, data buatan mengandung tambahan noise acak yang menyebarkan energi ke beberapa mode, sehingga dominasi Mode 1 menjadi lebih rendah. Hal ini menegaskan bahwa pada data observasi yang merepresentasikan kondisi nyata, ENSO memang merupakan pola variabilitas utama yang paling kuat.
</p>

<h3><b>2. Pola Spasial $U$</b></h3>
<p align="justify">
Matriks $U$ berisi vektor singular kiri $u_i$ yang merepresentasikan pola spasial dari setiap mode. Pola ini menunjukkan bagaimana anomali suhu permukaan laut terdistribusi secara geografis. Hasil simulasi memperlihatkan bahwa pola spasial Mode 1 menampilkan karakteristik pemanasan dan pendinginan di Samudra Pasifik bagian tengah-timur yang sesuai dengan definisi El Ni√±o dan La Ni√±a.
</p>
<p align="justify">
Secara sederhana, pola spasial ini dapat dipahami sebagai peta yang menunjukkan wilayah laut mana yang paling berperan dalam menentukan terjadinya El Ni√±o atau La Ni√±a. Wilayah yang memiliki nilai anomali terbesar merupakan area yang paling sensitif terhadap perubahan ENSO.
</p>

<h3><b>3. Pola Temporal $V^T$</b></h3>
<p align="justify">
Matriks $V^T$ berisi vektor singular kanan $v_i^T$ yang merepresentasikan variasi temporal dari masing-masing mode. Untuk interpretasi yang lebih bermakna, vektor ini dikalikan dengan nilai singular yang sesuai sehingga menunjukkan kekuatan pola spasial dari waktu ke waktu. Hasil simulasi menunjukkan bahwa pola temporal Mode 1 mengalami fluktuasi yang mencerminkan fase El Ni√±o (nilai positif) dan La Ni√±a (nilai negatif) selama periode 2014‚Äì2024.
</p>
<p align="justify">
Dalam konteks yang lebih mudah dipahami, grafik ini dapat dibaca sebagai informasi mengenai kapan ENSO menguat atau melemah, serta periode ketika El Ni√±o atau La Ni√±a mendominasi.
</p>

<h3><b>4. Spektrum Frekuensi</b></h3>
<p align="justify">
Analisis spektrum frekuensi dilakukan menggunakan Transformasi Fourier Cepat (FFT) terhadap komponen temporal Mode 1. Spektrum ini menunjukkan periode dominan dari variasi ENSO, yaitu seberapa sering pola tersebut berulang. Hasil analisis menunjukkan adanya puncak spektrum pada rentang periode sekitar 2 hingga 7 tahun, yang merupakan rentang periode khas ENSO sebagaimana dilaporkan dalam literatur klimatologi.
</p>
<p align="justify">
Secara sederhana, hasil ini menunjukkan bahwa ENSO bukanlah fenomena yang terjadi secara acak, melainkan memiliki pola berulang dalam jangka waktu beberapa tahun, sehingga memungkinkan untuk dianalisis dan diprediksi secara lebih sistematis.
